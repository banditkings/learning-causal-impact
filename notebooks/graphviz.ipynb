```python
import graphviz as gr
```

```python
g = gr.Graph('graph')
g.edges([('regular_sales_units', 'total_units'),
         ('promoted_sales_units_without_display', 'total_units'),
         ('promoted_sales_units_with_display', 'total_units')])
g.edges([('promoted_revenue_without_display', 'total_promoted_revenue'),
         ('promoted_revenue_with_display', 'total_promoted_revenue')])
g.edges([('display_weeks', 'promoted_revenue_with_display'),
         ('display_weeks', 'promoted_sales_units_with_display')])
g.edges([('promo_weeks', 'promoted_revenue_without_display'),
         ('promo_weeks', 'promoted_sales_units_without_display')])
g.edges([('total_gross_revenue', 'total_net_revenue'),
          ('promotional_spending', 'total_net_revenue'),
          ('regular_revenue', 'total_net_revenue'),
          ('total_promoted_revenue', 'total_net_revenue')])
g.edges([('regular_revenue', 'total_gross_revenue'),
         ('promoted_sales_units_without_display', 'total_gross_revenue'),
         ('promoted_sales_units_with_display', 'total_gross_revenue')])
g.edge('regular_sales_units', 'regular_revenue')
g.edge('total_net_revenue', 'total_gross_profit')
g.edge('total_cost', 'total_gross_profit')
g.edge('total_units', 'total_cost')
g.edge('PRODUCT_MFG_COST', 'total_cost')

g
```
